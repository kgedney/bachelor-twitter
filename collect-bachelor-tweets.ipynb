### 2 - Set up Tweepy Streaming Listener

```python
# tweepy streaming
# https://tweepy.readthedocs.io/en/3.7.0/streaming_how_to.html
# ref: https://www.dataquest.io/blog/streaming-data-python/

# 1. Create a class inheriting from StreamListener
# 2. Using that class create a Stream object
# 3. Connect to the Twitter API using the Stream.
```

```python
# update so collects in a nice format
class StreamListener(tweepy.StreamListener):

    def on_status(self, status):
        print(status.text)
        
        # create file to collect tweets json
        with open('tweets.json', 'a') as f:
            f.write(json.dumps(status._json))
            
    def on_error(self, status_code):
        print('Error')
        if status_code == 420:
            return False 
```

### 3 - Run Listener ! need to add time limit here

```python
# run
stream_listener = StreamListener()
stream = tweepy.Stream(auth=api.auth, listener=stream_listener)
stream.filter(track=["the bachelor", "#thebachelor", 'bachelor'])
```
